## INTRO 

This notebook will be a quick analysis on a movie review dataset, located [here](https://fivethirtyeight.com/features/fandango-movies-ratings/)

The dataset includes normalized reviews for movies from various sites like metacritic, rotten tomatoes, IMDB and fandango.  

Initialy, the dataset was compilled to see if fandango and other review sites may be inflating reviews so that people buy more movie tickets.  This notebook will explore some aspects of that possibility using basic statistics like histograms, correlation coefficients and linear regression

```python
import pandas as pd

data = pd.read_csv("fandango_score_comparison.csv")

data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(data['Fandango_Stars'])
plt.show()
plt.hist(data['Metacritic_norm_round'])
plt.show()
```

## Fandango vs Metacritic

Interestingly, there are no scores below a 3 on fandango... Let's take a closer look at the mean, median, and standard deviation.

```python
fan_mean = data['Fandango_Stars'].mean()
print("Fandango Mean:",fan_mean)
meta_mean = data['Metacritic_norm_round'].mean()
print("Metacritic Mean:",meta_mean)

fan_med = data['Fandango_Stars'].median()
print("\nFandango Median:", fan_med)
meta_med = data['Metacritic_norm_round'].median()
print("Metacritic Median:", meta_med)

fan_stddev = data['Fandango_Stars'].std()
print("\nFandango Standard Deviation:", fan_stddev)
meta_stddev = data['Metacritic_norm_round'].std()
print("Metacricit Standard Deviation:", meta_stddev)
```

Both the mean and median are skewed a whole point higher for fandango.  Also considering the standard deviation is significantly smaller for fandango, their reviews are definately more congregated towards the top end of the scale.  Metacritic is more spread out, and more similar to a normal distribution.  Which one would expect for movie reviews.  (Except it seems on all review scales the lower end of the scale is rarely used..)

This could be suggesting that fandango pushes their reviews up towards the top of the scale.  Since they also sell movie tickets, they could be incentivized  for doing so.  (This was the hypothese that the folks at fivethiryeight had when they cleaned this dataset in the first place) 

## Difference in Review Style

Metacritic  aggregates movie reviews, then sums them up to a 0 - 100 point based scale.  Which has been normalized to a 0 - 5 system to make it easier to compare with other sites.  Fandango uses a star system, 0 - 5, where half stars are allowed.  It is worth noting it is a bit harder to see how fandango is aggregating and creating its final review.

## Finding Outliers

```python
plt.scatter(data["Metacritic_norm_round"], data["Fandango_Stars"])
plt.show()
```

```python
data['fm_diff'] = abs(data['Metacritic_norm_round'] - data['Fandango_Stars'])
data.sort_values(by="fm_diff", ascending=False, inplace=True)
data[['FILM','fm_diff']].head(5)
```

## Correlations

```python
from scipy.stats import linregress 

slope, intercept, r_value, p_value, std_err = linregress(data['Metacritic_norm_round'], data['Fandango_Stars'])
print("Pearson R:", r_value)

pred_3 = 3 * slope + intercept
print("Metacritic 3.0 prediction:", pred_3)
```

There's actually quite a low correlation between fandango and metacritic scores.  Which suggests that each reviews site simply reviews things quite differently in general.  Even if there is or is not review inflation going on with fandango.

## Finding Residuals

```python
pred_1 = 1 * slope + intercept
pred_5 = 5 * slope + intercept

plt.scatter(data["Metacritic_norm_round"], data["Fandango_Stars"])
plt.plot([1.0, 5.0], [pred_1, pred_5])
plt.xlim(1,5)
plt.show()
```
