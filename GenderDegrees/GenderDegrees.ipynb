```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16, 16))

col_flag=0
bool_flag=True

for sp in range(0,18): 
    ax = fig.add_subplot(6,3,sp+1)
    
    if(col_flag==0 and bool_flag==True):  
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[int(sp/3)])
        col_flag=1
        bool_flag=False
    
    elif(col_flag==1 and bool_flag==True):
        
        if sp != 16:
            ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
            ax.set_title(lib_arts_cats[int(sp/3)])
       
        else:
            ax.set_visible(False)
        col_flag=2
    
    elif(col_flag==2 and bool_flag==True):
        ax.plot(women_degrees['Year'], women_degrees[other_cats[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[int(sp/3)])
        col_flag=0
    
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    bool_flag=True  
    
    if sp == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2007, 8, 'Men')
    
    elif sp == 1:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
        
    elif sp == 2:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
        
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom='on')
        
    elif sp == 17:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom='on')


plt.savefig('gender_degrees.png')
plt.show()
```
