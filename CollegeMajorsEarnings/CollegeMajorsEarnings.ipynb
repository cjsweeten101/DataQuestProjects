## Introduction:  Visualizing College Majors

In this notebook, I will be using the pandas wrapper for matplotlib to quickly analyze and visualize the dataset FiveThirtyEight cleaned and organized on the outcomes of students of various majors.  Located [here](https://github.com/fivethirtyeight/data/tree/master/college-majors)

The dataset itself contains columns on things like major, number of people in the major, gender ratios, employment status etc. . .

Some potential things to explore:

* Do more popular majors result in higher paying jobs?

* What majors are mostly male?  Which are mostly female?

* What types of majors have the most students?

Before any of that though, some basic cleaning is required.

```python
import pandas as pd
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
recent_grads.head()
recent_grads.tail()
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape

raw_data_count
cleaned_data_count
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size vs. Median')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size vs. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full Time vs Median')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind ='scatter')
ax.set_title('Share of Women vs Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind ='scatter')
ax.set_title('Men vs Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind ='scatter')
ax.set_title('Women and Median')
```

## Scatter Exploration

* First, it seems like there is no correlation between more popular majors (sample size) and more money (median).

* Also no correlation between female percentage and median income or employment rate

* Similarly, no correlation between number of full time employees and median income.

It interesting to note however, that the greatest varience in most of these plots occurs on the left end, where the sample size is smaller, or the number of full time employees is less.

If I were to wager a guess, this may be due to the so called "Law of Small Numbers".  When you have a significantly small sample size, it is much more likely to get more fluctuations.

```python
col_names = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(10,24))
for i in range(1,8):
    ax = fig.add_subplot(8,1,i)
    ax = recent_grads[col_names[i]].hist(bins=25)
    ax.set_title(col_names[i])
```

## Histogram Exploration

* There are similar percentages of majors that are predominantly male and female (Share women less then .2, and greater then .8)  And there seems to be a pretty large distribution around the .6/.7 range.


* The most common median salary range is around 3500, with the number of people earning greater then that dropping of pretty quickly.


* The unemployment peak is at 0.06, and is pretty close to a normal distribution.
